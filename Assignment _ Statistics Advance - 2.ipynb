## Statistics Part 2 Questions

### Q1. What is hypothesis testing in statistics?

--> Hypothesis testing is a statistical method used to check if an assumption about a population (called the **null hypothesis**) is likely to be true, based on sample data. It helps in making decisions using probabilities.

**Example:**  
A company claims that their battery lasts **10 hours**. A sample test shows an average life of **9.5 hours**. Hypothesis testing is used to check if this difference is significant or just due to random variation.




### Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

--> The **null hypothesis (H₀)** is a statement that assumes there is **no effect or no difference** in a population. It represents the default or original claim.

--> The **alternative hypothesis (H₁ or Ha)** is a statement that contradicts the null hypothesis. It suggests that **there is an effect or a difference**.

**Difference:**  
- **Null Hypothesis (H₀):** No change, no difference (status quo).  
- **Alternative Hypothesis (H₁):** There is a change or difference (new claim).

**Example:**  
H₀: The average test score is 70.  
H₁: The average test score is not 70.


### Q3. What is the significance level in hypothesis testing, and why is it important?

--> The **significance level** (denoted as **α**) is the probability of **rejecting the null hypothesis when it is actually true**. It is the threshold for deciding whether an observed result is statistically significant.

--> Common values of α are **0.05 (5%)**, **0.01 (1%)**, etc.

**Importance:**  
It helps control the risk of making a **Type I error** (false positive), ensuring our conclusions are reliable.

**Example:**  
If α = 0.05, we are willing to accept a 5% chance of wrongly rejecting a true null hypothesis.


### Q4. What does a P-value represent in hypothesis testing?

--> The **P-value** is the probability of getting results **as extreme as or more extreme than** the observed results, assuming the **null hypothesis is true**.

--> It helps us decide whether to reject the null hypothesis.

**Interpretation:**  
- A **small P-value** (≤ α) → Strong evidence **against** H₀ → **Reject** the null hypothesis.  
- A **large P-value** (> α) → Weak evidence against H₀ → **Fail to reject** the null hypothesis.

**Example:**  
If P-value = 0.03 and α = 0.05, we reject the null hypothesis because 0.03 < 0.05.


### Q5. How do you interpret the P-value in hypothesis testing?

--> The **P-value** tells us how likely the observed data would be if the **null hypothesis (H₀)** were true.

**Interpretation:**  
- **P-value ≤ α (significance level):** Reject H₀ → Result is **statistically significant**.  
- **P-value > α:** Fail to reject H₀ → Result is **not statistically significant**.

**Example:**  
If P-value = 0.02 and α = 0.05, we reject the null hypothesis. This means there is **strong evidence** that the effect or difference is real.


### 6. What are Type 1 and Type 2 errors in hypothesis testing?

--> In hypothesis testing, errors can occur when making a decision based on sample data.

**Type 1 Error (False Positive):**  
Rejecting the **null hypothesis (H₀)** when it is actually **true**.  
- Controlled by the **significance level α**.

**Type 2 Error (False Negative):**  
Failing to reject the **null hypothesis (H₀)** when it is actually **false**.  
- Related to the **power** of the test.

**Example:**  
- Type 1: Declaring a medicine works when it doesn’t.  
- Type 2: Saying a medicine doesn’t work when it actually does.


### Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

--> **One-tailed test:**  
Tests for an effect in **only one direction** (either greater than or less than).  
- Example: Testing if a new drug is **better** than the old one.

--> **Two-tailed test:**  
Tests for an effect in **both directions** (either greater or smaller).  
- Example: Testing if a new drug is **different** (better or worse) than the old one.

**Key difference:**  
One-tailed looks for change in one direction; two-tailed looks for any change.


### Q8. What is the Z-test, and when is it used in hypothesis testing?

--> The **Z-test** is a statistical test used to determine if there is a significant difference between sample and population means when the population variance is known and the sample size is large (usually > 30).

--> It assumes data is normally distributed.

**When used:**  
- Testing the mean of a large sample with known population variance.  
- Comparing two population means with large samples.

**Example:**  
Testing if the average height of students differs from the national average when population variance is known.


### Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

--> The **Z-score** measures how many standard deviations a data point or sample mean is from the population mean.

**Calculation:**  
\[
Z = \frac{(X - \mu)}{\sigma / \sqrt{n}}
\]  
Where:  
- \(X\) = sample mean  
- \(\mu\) = population mean  
- \(\sigma\) = population standard deviation  
- \(n\) = sample size

**Represents:**  
How far the sample mean is from the population mean in terms of standard errors, helping decide if the difference is significant.

**Example:**  
If Z = 2, the sample mean is 2 standard errors above the population mean.


### Q10. What is the T-distribution, and when should it be used instead of the normal distribution?

--> The **T-distribution** is a probability distribution similar to the normal distribution but with heavier tails. It accounts for extra uncertainty when estimating the population standard deviation from a small sample.

--> It should be used when:  
- The sample size is **small** (usually less than 30).  
- The population standard deviation is **unknown**.

**Example:**  
When testing the mean of a small sample of students’ test scores without knowing the population variance, use the T-distribution.


### Q11. What is the difference between a Z-test and a T-test?

| Aspect               | Z-test                          | T-test                            |
|----------------------|--------------------------------|----------------------------------|
| Population variance   | Known                          | Unknown                          |
| Sample size          | Large (usually > 30)            | Small (usually ≤ 30)             |
| Distribution used     | Normal (Z-distribution)         | T-distribution (with degrees of freedom) |
| Application           | Testing means with known variance | Testing means with unknown variance |

**Example:**  
- Use Z-test to check average height when population variance is known.  
- Use T-test to check average weight from a small sample when variance is unknown.


### Q12. What is the T-test, and how is it used in hypothesis testing?

--> The **T-test** is a statistical test used to compare the sample mean to the population mean (or compare means of two samples) when the population variance is **unknown** and the sample size is **small**.

--> It uses the **T-distribution** to determine if the difference is statistically significant.

**Usage:**  
- Testing if a sample mean differs from a known value.  
- Comparing means of two independent or paired samples.

**Example:**  
Testing if the average score of a small class differs from the national average when population variance is unknown.


### Q13. What is the relationship between Z-test and T-test in hypothesis testing?

--> Both **Z-test** and **T-test** are used to test hypotheses about population means.

--> The **main relationship** is:  
- When the **sample size is large** and **population variance is known**, use the **Z-test**.  
- When the **sample size is small** or **population variance is unknown**, use the **T-test**.

--> As sample size increases, the T-distribution approaches the normal distribution, so the **T-test becomes similar to the Z-test**.

**Example:**  
For large samples, T-test results approximate Z-test results.


### Q14. What is a confidence interval, and how is it used to interpret statistical results?

--> A **confidence interval (CI)** is a range of values, calculated from sample data, that is likely to contain the true population parameter (like the mean) with a certain level of confidence (e.g., 95%).

--> It shows the **uncertainty** around the sample estimate.

**Usage:**  
- If a 95% CI for a mean is (45, 55), we are 95% confident that the true population mean lies between 45 and 55.  
- Helps in making decisions without just relying on a single point estimate.

**Example:**  
A 95% CI for average height is (160 cm, 170 cm), meaning the true average height is likely in this range.


### Q15. What is the margin of error, and how does it affect the confidence interval?

--> The **margin of error (MOE)** is the maximum amount by which the sample estimate is expected to differ from the true population parameter.

--> It defines the **half-width** of the confidence interval.

**Effect on Confidence Interval:**  
- Confidence Interval = Sample Estimate ± Margin of Error  
- A larger margin of error means a wider confidence interval, indicating less precision.  
- A smaller margin of error means a narrower confidence interval, indicating more precision.

**Example:**  
If the sample mean is 50 and MOE is 5, the 95% confidence interval is (45, 55).


### Q16. How is Bayes' Theorem used in statistics, and what is its significance?

--> **Bayes' Theorem** calculates the probability of an event based on prior knowledge of conditions related to the event.

--> It updates the probability of a hypothesis as new evidence or data becomes available.

**Significance:**  
- Helps in **revising predictions** or beliefs with new data (called **posterior probability**).  
- Widely used in decision making, machine learning, and medical testing.

**Example:**  
Given a positive medical test result, Bayes’ Theorem helps find the probability that the patient actually has the disease.


### Q17. What is the Chi-square distribution, and when is it used?

--> The **Chi-square distribution** is a probability distribution used for testing relationships between categorical variables.

--> It is the distribution of a sum of the squares of independent standard normal random variables.

**When used:**  
- In **Chi-square tests** for independence or goodness-of-fit.  
- To check if observed data fits expected distributions.  
- To test association between two categorical variables.

**Example:**  
Testing if gender (male/female) is independent of preference for a product.


### Q18. What is the Chi-square goodness of fit test, and how is it applied?

--> The **Chi-square goodness of fit test** checks if observed categorical data matches an expected distribution.

--> It compares the observed frequencies with expected frequencies under a specific hypothesis.

**How it’s applied:**  
1. State the expected distribution.  
2. Collect observed data counts.  
3. Calculate the Chi-square statistic:  
\[
\chi^2 = \sum \frac{(O - E)^2}{E}
\]  
where \(O\) = observed frequency, \(E\) = expected frequency.  
4. Compare the statistic to the Chi-square distribution with appropriate degrees of freedom to decide if data fits the expected distribution.

**Example:**  
Testing if a dice is fair by comparing observed roll counts to expected equal probabilities.


### Q19. What is the F-distribution, and when is it used in hypothesis testing?

--> The **F-distribution** is a probability distribution used to compare variances and test if two population variances are equal.

--> It is the distribution of the ratio of two independent chi-square variables divided by their degrees of freedom.

**When used:**  
- In **ANOVA (Analysis of Variance)** to test differences between group means.  
- To test equality of variances (e.g., in the F-test).

**Example:**  
Testing if the variance in test scores is the same across different classes.


### Q20. What is an ANOVA test, and what are its assumptions?

--> **ANOVA (Analysis of Variance)** is a statistical test used to compare the means of **three or more groups** to see if at least one group mean is different.

**Assumptions:**  
1. **Independence:** Observations are independent.  
2. **Normality:** The data in each group is approximately normally distributed.  
3. **Homogeneity of variances:** The variances across groups are equal.

**Example:**  
Comparing average exam scores of students from three different schools to see if they differ.


### Q21. What are the different types of ANOVA tests?

1. **One-Way ANOVA:**  
   - Compares means of **one independent factor** with three or more groups.  
   - Example: Comparing test scores among different teaching methods.

2. **Two-Way ANOVA:**  
   - Compares means based on **two independent factors** and can test interaction effects between factors.  
   - Example: Comparing test scores based on teaching method and gender.

3. **Repeated Measures ANOVA:**  
   - Used when the **same subjects** are measured multiple times under different conditions.  
   - Example: Measuring blood pressure before and after treatment on the same patients.

Each type helps analyze data depending on the study design.


### Q22. What is the F-test, and how does it relate to hypothesis testing?

--> The **F-test** is a statistical test that compares the variances of two or more groups to see if they are significantly different.

--> It uses the **F-distribution** and is commonly used in hypothesis testing for:  
- Comparing group variances  
- Testing overall significance in **ANOVA**

**Relation to hypothesis testing:**  
- Null hypothesis (\(H_0\)): All group variances (or means) are equal.  
- Alternative hypothesis (\(H_a\)): At least one group variance (or mean) differs.  
- If the calculated F-statistic is greater than the critical value, reject \(H_0\).

**Example:**  
Using F-test to check if the average test scores differ across multiple classes.

```python
# Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate PDF of Bernoulli distribution.
    x: 0 or 1 (outcome)
    p: probability of success (1)
    """
    if x not in [0, 1]:
        return 0
    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.6  # Probability of success
print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

```python
# Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, data):
    mean = np.mean(data)
    std = np.std(data)
    return (data_point - mean) / std

# Sample data
data = np.random.normal(50, 10, 1000)
data_point = 65
z_score = calculate_z_score(data_point, data)

# Plot standard normal distribution
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, 0, 1)
plt.plot(x, pdf, label='Standard Normal PDF')

# Mark the Z-score on the plot
plt.axvline(z_score, color='r', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.title('Z-score Comparison with Standard Normal Distribution')
plt.legend()
plt.show()
```

```python
# Q15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Sample data
sample = np.array([52, 55, 51, 53, 54, 56, 50, 57, 52, 54])
sample_mean = np.mean(sample)
sample_size = len(sample)
population_mean = 50
population_std = 3

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Q16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

data = np.array([12, 15, 14, 16, 13, 14, 15, 16, 17, 14])
confidence_level = 0.95
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error

# Calculate confidence interval
h = std_err * stats.t.ppf((1 + confidence_level) / 2, n - 1)
lower_bound = mean - h
upper_bound = mean + h

print(f"Mean: {mean:.2f}")
print(f"{int(confidence_level*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation:
# We are 95% confident that the true population mean lies within this interval.
```

```python
# Q1. Perform a Z-test comparing a sample mean to a known population mean and interpret the result.

import numpy as np
from scipy.stats import norm

# Sample data
sample = np.array([52, 55, 48, 51, 54, 49, 53, 50, 52, 54])
sample_mean = np.mean(sample)
population_mean = 50
population_std = 5  # Known population std dev
n = len(sample)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject null hypothesis: No significant difference from population mean.")
```

```python
# Q2. Simulate random data for hypothesis testing and calculate the P-value.

import numpy as np
from scipy.stats import ttest_1samp

# Simulate sample data (e.g., sample from a population with mean=50)
np.random.seed(0)
sample = np.random.normal(loc=52, scale=5, size=30)

pop_mean = 50

t_stat, p_value = ttest_1samp(sample, pop_mean)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject null hypothesis: Sample mean differs significantly from population mean.")
else:
    print("Fail to reject null hypothesis: No significant difference found.")
```

```python
# Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sample = np.array([51, 53, 49, 52, 54, 50, 48, 55, 47, 53])
sample_mean = np.mean(sample)
pop_mean = 50
pop_std = 3
n = len(sample)

z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
alpha = 0.05
z_crit = norm.ppf(1 - alpha/2)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
plt.plot(x, y)
plt.fill_between(x, 0, y, where=(x <= -z_crit), color='red', alpha=0.3)
plt.fill_between(x, 0, y, where=(x >= z_crit), color='red', alpha=0.3)
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z = {z_score:.2f}')
plt.title('Two-tailed Z-test Decision Regions')
plt.legend()
plt.show()

if abs(z_score) > z_crit:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
# Q5. Calculate and visualize Type 1 and Type 2 errors in hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type_errors(mu0=0, mu1=1, sigma=1, alpha=0.05):
    z_crit = norm.ppf(1 - alpha)  # Critical value for alpha

    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    null_dist = norm.pdf(x, mu0, sigma)  # Null distribution
    alt_dist = norm.pdf(x, mu1, sigma)   # Alternative distribution

    plt.plot(x, null_dist, label='Null Hypothesis', color='blue')
    plt.plot(x, alt_dist, label='Alternative Hypothesis', color='green')

    # Shade Type 1 error region
    plt.fill_between(x, 0, null_dist, where=(x > mu0 + z_crit*sigma), color='red', alpha=0.4, label='Type 1 Error (α)')
    # Shade Type 2 error region
    plt.fill_between(x, 0, alt_dist, where=(x <= mu0 + z_crit*sigma), color='orange', alpha=0.4, label='Type 2 Error (β)')

    plt.axvline(mu0 + z_crit*sigma, color='black', linestyle='--', label='Critical value')
    plt.title('Type 1 and Type 2 Errors')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

plot_type_errors()
```

```python
# Q6. Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy.stats import ttest_ind

# Sample data: two independent groups
group1 = np.array([23, 20, 22, 21, 19])
group2 = np.array([30, 29, 31, 32, 28])

# Perform independent t-test
t_stat, p_value = ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpret results
if p_value < 0.05:
    print("Reject null hypothesis: Groups have different means.")
else:
    print("Fail to reject null hypothesis: No significant difference between groups.")
```

```python
# Q8. Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy.stats import ttest_1samp, norm

np.random.seed(0)
data = np.random.normal(50, 10, 30)

pop_mean = 52
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

z_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

t_stat, p_value_t = ttest_1samp(data, pop_mean)

print(f"Sample mean: {sample_mean:.3f}")
print(f"Z-test statistic: {z_stat:.3f}, P-value: {p_value_z:.3f}")
print(f"T-test statistic: {t_stat:.3f}, P-value: {p_value_t:.3f}")

if p_value_z < 0.05:
    print("Z-test: Reject null hypothesis")
else:
    print("Z-test: Fail to reject null hypothesis")

if p_value_t < 0.05:
    print("T-test: Reject null hypothesis")
else:
    print("T-test: Fail to reject null hypothesis")
```

```python
# Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error
    margin = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    lower = mean - margin
    upper = mean + margin
    return lower, upper

# Significance:
# The confidence interval gives a range where the true population mean likely falls with a certain confidence level (e.g., 95%).

# Example usage:
np.random.seed(0)
sample_data = np.random.normal(100, 15, 50)
ci_lower, ci_upper = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = stats.sem(data)  # Standard error of the mean
    z_score = stats.t.ppf((1 + confidence) / 2, n - 1)  # t-score for confidence level
    margin = z_score * std_err
    return margin

# Example usage:
np.random.seed(1)
sample_data = np.random.normal(50, 8, 40)
me = margin_of_error(sample_data)
print(f"Margin of Error at 95% confidence: {me:.3f}")
```

```python
# Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

# Bayes' Theorem formula:
# P(A|B) = [P(B|A) * P(A)] / P(B)

def bayes_theorem(prior_A, likelihood_B_given_A, prior_notA, likelihood_B_given_notA):
    # Calculate total probability P(B)
    prob_B = likelihood_B_given_A * prior_A + likelihood_B_given_notA * prior_notA
    # Calculate posterior probability P(A|B)
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / prob_B
    return posterior_A_given_B

# Explanation:
# prior_A = P(A), initial belief about event A
# likelihood_B_given_A = P(B|A), probability of evidence B given A
# prior_notA = P(~A), initial belief about not A
# likelihood_B_given_notA = P(B|~A), probability of B given not A
# The function calculates updated belief P(A|B) after observing evidence B.

# Example:
# Disease testing:
# Prior probability of disease: 1%
# Probability of positive test if disease: 99%
# Probability of positive test if no disease: 5%

prior_disease = 0.01
likelihood_pos_given_disease = 0.99
prior_no_disease = 1 - prior_disease
likelihood_pos_given_no_disease = 0.05

posterior = bayes_theorem(prior_disease, likelihood_pos_given_disease, prior_no_disease, likelihood_pos_given_no_disease)
print(f"Posterior probability of disease given positive test: {posterior:.4f}")
```

```python
# Q12. Perform a Chi-square test for independence between two categorical variables in Python

import pandas as pd
from scipy.stats import chi2_contingency

data = {'Yes': [30, 10],
        'No': [20, 40]}
categories = ['Male', 'Female']

contingency_table = pd.DataFrame(data, index=categories)

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("P-value:", p)
print("Expected frequencies:\n", expected)

# If p < 0.05, reject null hypothesis → variables dependent
# Else, variables independent
```

```python
# Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np

# Observed frequency table
observed = np.array([[30, 10],
                     [20, 40]])

# Calculate row sums, column sums, and total
row_sums = observed.sum(axis=1).reshape(-1, 1)
col_sums = observed.sum(axis=0).reshape(1, -1)
total = observed.sum()

# Calculate expected frequencies
expected = (row_sums @ col_sums) / total

print("Observed Frequencies:\n", observed)
print("Expected Frequencies:\n", expected)
```

```python
# Q14. Perform a goodness-of-fit test using Python to compare observed data to an expected distribution

from scipy.stats import chisquare

observed = [50, 30, 20]
total_observed = sum(observed)

expected_proportions = [1/3, 1/3, 1/3]

expected = [p * total_observed for p in expected_proportions]

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)

# Interpretation:
# If p < 0.05, observed data significantly differ from expected distribution
# Otherwise, no significant difference
```
